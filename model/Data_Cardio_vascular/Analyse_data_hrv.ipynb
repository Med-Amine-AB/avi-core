```python
import pandas as pd
from sklearn.ensemble import IsolationForest
import joblib   # pour sauvegarder le modèle

# 1) Charger & préparer
df = pd.read_csv("Data_HRV_hr_rate.csv")
X   = df[["Average_Heart_Rate", "hrv"]]

# 2) Entraîner sur données 'normales'
iso = IsolationForest(
        n_estimators=200,
        contamination=0.05,
        random_state=42)
iso.fit(X)

```

```python
# 3) Sauvegarder (facultatif)
joblib.dump(iso, "iso_effort.pkl")
```

```python
import pandas as pd, joblib
iso = joblib.load("iso_effort.pkl")

sample = pd.DataFrame({
        "Average_Heart_Rate": [120],
        "hrv": [30]})

pred = iso.predict(sample)[0]      # +1 = normal, -1 = effort/anomaly
print("Résultat :", "Effort" if pred==-1 else "Normal")
```
