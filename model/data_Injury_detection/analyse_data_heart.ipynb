```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('H_R.csv', low_memory=False)
```

```python
df
```

```python
# Convert the 'Time' column to datetime
df['Time'] = pd.to_datetime(df['Time'])
```

```python
# Set the 'Time' column as the index
df.set_index('Time', inplace=True)
```

```python

# Resample the data to 1-minute intervals and calculate the mean heart rate
heart_rate_per_minute = df['Value'].resample('1min').mean()
```

```python

# Convert to DataFrame and reset index to make 'Time' a column
heart_rate_per_minute = heart_rate_per_minute.reset_index()
```

```python
# Rename columns to match desired output
heart_rate_per_minute.columns = ['Time', 'Average_Heart_Rate']
```

```python
# Round Average_Heart_Rate to 2 decimal places for consistency with example
heart_rate_per_minute['Average_Heart_Rate'] = heart_rate_per_minute['Average_Heart_Rate'].round(2)
```

```python
# Display the first few rows to match the example
print(heart_rate_per_minute.head(4))
```

```python
# Optionally, save the result to a CSV file
heart_rate_per_minute.to_csv('heart_rate_per_minute.csv', index=False)
```
