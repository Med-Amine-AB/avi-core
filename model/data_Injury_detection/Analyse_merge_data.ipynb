```python
import pandas as pd
import numpy as np
from sklearn.ensemble import IsolationForest, RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
import joblib

# Load the dataset
df = pd.read_csv('cleaned_merged_data_final.csv')

# Ensure 'Time' is in datetime format
df['Time'] = pd.to_datetime(df['Time'])

# Handle missing values
# First forward fill, then fill remaining NaNs with column means
df = df.ffill()  # This replaces the deprecated fillna(method='ffill')
numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())
```

```python
df
```

```python
# Feature Engineering
# 1. Acceleration magnitude
df['Accel_Magnitude'] = np.sqrt(df['X']**2 + df['Y']**2 + df['Z']**2)
```

```python

# 2. Acceleration change (difference between consecutive readings)
df['Accel_Change'] = df['Accel_Magnitude'].diff().fillna(0)
```

```python
# 3. Heart rate variability (standard deviation over a 5-minute window)
df['HR_Variability'] = df['Average_Heart_Rate'].rolling(window=5, min_periods=1).std().fillna(0)
```

```python
# 4. Temperature anomaly (deviation from mean)
temp_mean = df['TEMP'].mean()
df['Temp_Anomaly'] = df['TEMP'] - temp_mean
```

```python
df
```

```python

# Select features for modeling
features = ['X', 'Y', 'Z', 'Accel_Magnitude', 'Accel_Change', 'Average_Heart_Rate', 'HR_Variability', 'TEMP', 'Temp_Anomaly']
X = df[features]
```

```python
# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# --- Unsupervised Model: Isolation Forest ---
# Train Isolation Forest for anomaly detection
iso_forest = IsolationForest(contamination=0.01, random_state=42)  # Assume 1% of data are anomalies
df['Anomaly_Score'] = iso_forest.fit_predict(X_scaled)
df['Is_Anomaly'] = df['Anomaly_Score'] == -1  # -1 indicates anomaly
```

```python

# Display detected anomalies
anomalies = df[df['Is_Anomaly']][['Time', 'X', 'Y', 'Z', 'Average_Heart_Rate', 'TEMP', 'Accel_Magnitude', 'HR_Variability', 'Temp_Anomaly']]
print("Detected Anomalies (Isolation Forest):")
print(anomalies)
```

```python

# --- Supervised Model: Random Forest (Alternative, if labels available) ---
# For demonstration, assume a synthetic 'Injury' column (0 = no injury, 1 = injury)
# Replace this with actual labels if available
np.random.seed(42)
df['Injury'] = np.random.choice([0, 1], size=len(df), p=[0.99, 0.01])  # 1% injury for demo
```

```python
# Prepare data for supervised learning
y = df['Injury']
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Train Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
```

```python
# Evaluate the model
y_pred = rf_model.predict(X_test)
print("\nRandom Forest Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
# Save the Random Forest model
joblib.dump(rf_model, 'random_forest_model.pkl')
print("Random Forest model saved as 'random_forest_model.pkl'")
```

```python
# Optional: Plot anomalies (requires matplotlib)
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.plot(df['Time'], df['Average_Heart_Rate'], label='Heart Rate', alpha=0.5)
plt.scatter(df[df['Is_Anomaly']]['Time'], df[df['Is_Anomaly']]['Average_Heart_Rate'], 
            color='red', label='Anomalies', marker='x')
plt.xlabel('Time')
plt.ylabel('Average Heart Rate (bpm)')
plt.title('Heart Rate with Detected Anomalies')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('anomaly_plot.png')
print("Anomaly plot saved as 'anomaly_plot.png'")
```
